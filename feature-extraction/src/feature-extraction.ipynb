# Feature extraction using SIFT, SURF, BRIEF and ORB.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import imutils
```

```python
input_img = ['Alexandra_Vodjanikova_0001.jpg', 'Alexandra_Vodjanikova_0002.jpg', 'Alexandra_Pelosi_0001.jpg', 'Alexandra_Rozovskaya_0001.jpg', 'Alexandra_Spann_0001.jpg']
```

## SIFT

```python
def plot_cv_img(input_image, output_image1, output_image2, output_image3):
    fig, ax = plt.subplots(nrows=2, ncols=2)
    ax[0][0].imshow(input_image, cmap = 'gray')
    ax[0][0].set_title('Input Image')
    ax[0][0].axis('off')
    ax[0][1].imshow(output_image1, cmap = 'gray')
    ax[0][1].set_title('SIFT Gray')
    ax[0][1].axis('off')
    ax[1][0].imshow(output_image2, cmap = 'gray')
    ax[1][0].set_title('SIFT Color')
    ax[1][0].axis('off')
    ax[1][1].imshow(output_image3, cmap = 'gray')
    ax[1][1].set_title('SIFT Rotated')
    ax[1][1].axis('off')
    plt.show()
```

```python
i = 0
for path in input_img:
    i = i + 1
    img = cv2.imread(path)
    c_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    r_img = imutils.rotate(c_img, angle=10)
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    sift = cv2.xfeatures2d.SIFT_create()
    kp, des = sift.detectAndCompute(gray_img, None)
    kp_c, des_c = sift.detectAndCompute(c_img, None)
    kp_r, des_r = sift.detectAndCompute(r_img, None)

    kp_img_g = cv2.drawKeypoints(gray_img, kp, None, color=(0, 255, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    kp_img_c = cv2.drawKeypoints(c_img, kp_c, None, color=(0, 255, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    kp_img_r = cv2.drawKeypoints(r_img, kp_r, None, color=(0, 255, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    print("\nCase -", i, "\n")
    plot_cv_img(c_img , kp_img_g, kp_img_c, kp_img_r)
```

**Remarks:**
- We can see that in all the cases above, some more features are detected along the edges but nothing has been changes in the middle of the images when we rotate an image by some degree (10 in this case) and apply SIFT on that image . 
- Most of the time more features are extracted in color image than in that in the greyscale image. _For example, in the case - 3 of the above cases, in the upper right of the greyscale image in the hairs only three features have been identified while in the color image 4 distinct features have been identified._

## BRIEF

```python
def plot_cv_img(input_image, output_image1, output_image2, output_image3):
    fig, ax = plt.subplots(nrows=2, ncols=2)
    ax[0][0].imshow(input_image, cmap = 'gray')
    ax[0][0].set_title('Input Image')
    ax[0][0].axis('off')
    ax[0][1].imshow(output_image1, cmap = 'gray')
    ax[0][1].set_title('BRIEF Gray')
    ax[0][1].axis('off')
    ax[1][0].imshow(output_image2, cmap = 'gray')
    ax[1][0].set_title('BRIEF Color')
    ax[1][0].axis('off')
    ax[1][1].imshow(output_image3, cmap = 'gray')
    ax[1][1].set_title('BRIEF Rotated')
    ax[1][1].axis('off')
    plt.show()
```

## ORB

```python
def plot_cv_img(input_image, output_image1, output_image2, output_image3):
    fig, ax = plt.subplots(nrows=2, ncols=2)
    ax[0][0].imshow(input_image, cmap = 'gray')
    ax[0][0].set_title('Input Image')
    ax[0][0].axis('off')
    ax[0][1].imshow(output_image1, cmap = 'gray')
    ax[0][1].set_title('ORB Gray')
    ax[0][1].axis('off')
    ax[1][0].imshow(output_image2, cmap = 'gray')
    ax[1][0].set_title('ORB Color')
    ax[1][0].axis('off')
    ax[1][1].imshow(output_image3, cmap = 'gray')
    ax[1][1].set_title('ORB Rotated')
    ax[1][1].axis('off')
    plt.show()
```

```python
i = 0
for path in input_img:
    i = i + 1
    img = cv2.imread(path)
    c_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    r_img = imutils.rotate(c_img, angle=10)
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    orb = cv2.ORB_create(nfeatures=2000)
    kp, des = orb.detectAndCompute(gray_img, None)
    kp_c, des_c = orb.detectAndCompute(c_img, None)
    kp_r, des_r = orb.detectAndCompute(r_img, None)

    kp_img_g = cv2.drawKeypoints(gray_img, kp, None, color=(0, 255, 0), flags=0)
    kp_img_c = cv2.drawKeypoints(c_img, kp_c, None, color=(0, 255, 0), flags=0)
    kp_img_r = cv2.drawKeypoints(r_img, kp_r, None, color=(0, 255, 0), flags=0)
    print("\nCase -", i, "\n")
    plot_cv_img(c_img , kp_img_g, kp_img_c, kp_img_r)
```

**Remarks:**
-  Most of the time, the images are similar when ORB is applied on images no-matter whether it is greyscale or colored image or rotated.
